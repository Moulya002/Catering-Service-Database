# Connect to sql server

```python
# Connect to MySQL server
db_connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='#Group12'
)
cursor = db_connection.cursor()

# Create the database
database_name = "CateringServices"
cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database_name}")
cursor.execute(f"USE {database_name}")
```

## Code to create Role table

```python
# Use the database
try:
    cursor.execute(f"USE {database_name}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit(1)

# SQL Statement to create table
create_role_Table = """
    CREATE TABLE IF NOT EXISTS Role (
        roleID INT AUTO_INCREMENT PRIMARY KEY,
        roleName VARCHAR(45) NOT NULL UNIQUE,
        canAssignEvents BOOLEAN NOT NULL,
        canDistributeCoupons BOOLEAN NOT NULL,
        canViewSalaries BOOLEAN NOT NULL,
        canManageInventory BOOLEAN NOT NULL,
        canScheduleEvents BOOLEAN NOT NULL
    )
    """

try:
    # Execute the insert query
    cursor.execute(create_role_Table)
    # Commit the transaction
    db_connection.commit()
    print("Role Table created successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    db_connection.rollback()
```

## Code to create Employee table

```python
# Use the database
try:
    cursor.execute(f"USE {database_name}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit(1)

# SQL Statement to create table
create_employee_Table = """
    CREATE TABLE IF NOT EXISTS Employee (
        employeeID INT AUTO_INCREMENT PRIMARY KEY,
        firstName VARCHAR(45) NOT NULL,
        middleName VARCHAR(45),
        lastName VARCHAR(45) NOT NULL,
        phoneNum VARCHAR(15) NOT NULL UNIQUE,
        emailID VARCHAR(50) NOT NULL,
        aptNum VARCHAR(15),
        streetName VARCHAR(45),
        city VARCHAR(30),
        state VARCHAR(30),
        pincode INT,
        gender VARCHAR(10),
        employmentType VARCHAR(30),
        roleID INT NOT NULL,
        FOREIGN KEY (roleID) REFERENCES Role(roleID)
    )
    """

try:
    # Execute the insert query
    cursor.execute(create_employee_Table)
    # Commit the transaction
    db_connection.commit()
    print("Employee Table created successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    db_connection.rollback()
```

## Code to create Location table

```python
# Use the database
try:
    cursor.execute(f"USE {database_name}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit(1)

# SQL Statement to create table
create_location_Table = """
    CREATE TABLE IF NOT EXISTS Location (
        locationID INT AUTO_INCREMENT PRIMARY KEY,
        locationName VARCHAR(45) NOT NULL,
        streetName VARCHAR(45),
        city VARCHAR(30),
        state VARCHAR(30),
        pincode INT,
        capacity INT NOT NULL,
        isAvailable VARCHAR(10),
        availabilityNotes VARCHAR(100),
        dailyEventLimit INT,
        lastUpdated DATETIME,
        locationType VARCHAR(50)
    )
    """

try:
    # Execute the insert query
    cursor.execute(create_location_Table)
    # Commit the transaction
    db_connection.commit()
    print("Location Table created successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    db_connection.rollback()
```

## Code to create Events table

```python
# Use the database
try:
    cursor.execute(f"USE {database_name}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit(1)

# SQL Statement to create table
create_events_Table = """
    CREATE TABLE IF NOT EXISTS Events (
        eventID INT AUTO_INCREMENT PRIMARY KEY,
        eventName VARCHAR(100) NOT NULL,
        locationID INT NOT NULL,
        eventTime DATETIME NOT NULL,
        eventHours INT NOT NULL,
        handledByManagerID INT,
        supervisorID INT,
        approved VARCHAR(10) ,
        FOREIGN KEY (locationID) REFERENCES Location(locationID),
        FOREIGN KEY (handledByManagerID) REFERENCES Employee(employeeID),
        FOREIGN KEY (supervisorID) REFERENCES Employee(employeeID)
    )
    """

try:
    # Execute the insert query
    cursor.execute(create_events_Table)
    # Commit the transaction
    db_connection.commit()
    print("Events Table created successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    db_connection.rollback()
```

## Code to create Coupons table

```python
# Use the database
try:
    cursor.execute(f"USE {database_name}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit(1)

# SQL Statement to create table
create_coupons_Table = """
    CREATE TABLE IF NOT EXISTS Coupons (
        couponID INT AUTO_INCREMENT PRIMARY KEY,
        employeeID INT NOT NULL,
        couponValue INT NOT NULL,
        redemptionStatus VARCHAR(15),
        allocatedByEmployeeID INT,
        FOREIGN KEY (employeeID) REFERENCES Employee(employeeID),
        FOREIGN KEY (allocatedByEmployeeID) REFERENCES Employee(employeeID)
    )
    """

try:
    # Execute the insert query
    cursor.execute(create_coupons_Table)
    # Commit the transaction
    db_connection.commit()
    print("Coupons Table created successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    db_connection.rollback()
```

## Code to create Salaries table

```python
# Use the database
try:
    cursor.execute(f"USE {database_name}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit(1)

# SQL Statement to create the Salaries table
create_salaries_Table = """
    CREATE TABLE IF NOT EXISTS Salaries (
        salaryID INT AUTO_INCREMENT PRIMARY KEY,  -- Unique identifier for salary records
        employeeID INT NOT NULL,                  -- Foreign key referencing Employee table
        salaryAmount INT,
        hoursWorked INT,
        hourlyPay INT,
        fixedPay INT,
        monthYear VARCHAR(7),                     -- Tracks salary period (e.g., '2024-11')
        UNIQUE (employeeID, monthYear),           -- Ensures each employee gets one salary entry per month
        FOREIGN KEY (employeeID) REFERENCES Employee(employeeID)
    )
"""

try:
    # Execute the create table query
    cursor.execute(create_salaries_Table)
    # Commit the transaction
    db_connection.commit()
    print("Salaries Table created successfully with unique employeeID and monthYear constraint!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    db_connection.rollback()
```

## Code to create Inventory table

```python
# Use the database
try:
    cursor.execute(f"USE {database_name}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit(1)

# SQL Statement to create table
create_inventory_table = """
    CREATE TABLE IF NOT EXISTS Inventory (
        inventoryID INT AUTO_INCREMENT PRIMARY KEY,
        itemName VARCHAR(100) NOT NULL,
        itemCategory VARCHAR(50),
        quantity INT NOT NULL,
        vendorID INT NOT NULL,
        eventID INT,
        FOREIGN KEY (vendorID) REFERENCES Vendor(vendorID),
        FOREIGN KEY (eventID) REFERENCES Events(eventID)
    )
    """

try:
    # Execute the insert query
    cursor.execute(create_inventory_table)
    # Commit the transaction
    db_connection.commit()
    print("Inventory Table created successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    db_connection.rollback()
```

## This query calculates the total salary paid to employees in December 2024, grouped by their employment type (e.g., Full-Time, Part-Time).

```python
# Query to calculate total salary paid by employment type for December 2024
query = """
SELECT
    e.employmentType,
    SUM(s.salaryAmount) AS totalSalaryPaid
FROM
    Employee e
JOIN
    Salaries s ON e.employeeID = s.employeeID
WHERE
    s.monthYear = '2024-12'
GROUP BY
    e.employmentType;
"""

try:
    # Execute the query
    cursor.execute(query)

    # Fetch all results
    results = cursor.fetchall()

    # Print the results
    print("Total Salary Paid by Employment Type for the month:")
    for row in results:
        employment_type = row[0]
        total_salary_paid = row[1]
        print(f"Employment Type: {employment_type}, Total Salary Paid: {total_salary_paid:.2f}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
```

## This query counts the number of events hosted at each location and orders the results by the total number of events in descending order (most events first).

```python
query = """
SELECT
    l.locationName,
    COUNT(e.eventID) AS eventCount
FROM
    Location l
JOIN
    Events e ON l.locationID = e.locationID
GROUP BY
    l.locationName
ORDER BY
    eventCount DESC;
"""

cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(f"Location Name: {row[0]}, Event Count: {row[1]}")
```

## This query identifies the top 5 employees who have the highest number of event assignments and lists their first name, last name, and total number of assignments.

```python
query = """
SELECT
    e.firstName,
    e.lastName,
    COUNT(eea.assignmentID) AS totalAssignments
FROM
    Employee e
JOIN
    EmployeeEventAssignments eea ON e.employeeID = eea.employeeID
GROUP BY
    e.employeeID
ORDER BY
    totalAssignments DESC
LIMIT 5;
"""

cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(f"Employee: {row[0]} {row[1]}, Total Assignments: {row[2]}")
```

## This query retrieves the top 5 employees with the highest number of event assignments, along with their first name, last name, total assignments, and a list of the names of the events they worked on.

```python
query = """
SELECT
    e.firstName,
    e.lastName,
    COUNT(eea.assignmentID) AS totalAssignments,
    GROUP_CONCAT(ev.eventName SEPARATOR ', ') AS eventNames
FROM
    Employee e
JOIN
    EmployeeEventAssignments eea ON e.employeeID = eea.employeeID
JOIN
    Events ev ON eea.eventID = ev.eventID
GROUP BY
    e.employeeID
ORDER BY
    totalAssignments DESC
LIMIT 5;
"""

try:
    # Execute the query
    cursor.execute(query)

    # Fetch all results
    results = cursor.fetchall()

    # Print the results
    print("Top 5 Employees by Total Assignments with Event Details:")
    for row in results:
        employee_name = f"{row[0]} {row[1]}"
        total_assignments = row[2]
        event_names = row[3]
        print(f"Employee: {employee_name}, Total Assignments: {total_assignments}, Events: {event_names}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
```

## This query retrieves information about the total number of coupons issued to each employee and how many of those coupons have been redeemed.

```python
query = """
SELECT
    e.firstName,
    e.lastName,
    COUNT(c.couponID) AS totalCouponsIssued,
    SUM(CASE WHEN c.redemptionStatus = 'Redeemed' THEN 1 ELSE 0 END) AS totalCouponsRedeemed
FROM
    Employee e
JOIN
    Coupons c ON e.employeeID = c.employeeID
GROUP BY
    e.employeeID;
"""

cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(f"Employee: {row[0]} {row[1]}, Total Coupons Issued: {row[2]}, Total Coupons Redeemed: {row[3]}")
```

## This query retrieves the first name and last name of employees who are assigned to events taking place at the location with the maximum capacity.

```python
# Define the subquery
query = """
SELECT firstName, lastName
FROM Employee
WHERE employeeID IN (
    SELECT employeeID
    FROM EmployeeEventAssignments
    WHERE eventID IN (
        SELECT eventID
        FROM Events
        WHERE locationID = (
            SELECT locationID
            FROM Location
            WHERE capacity = (SELECT MAX(capacity) FROM Location)
        )
    )
);
"""

try:
    # Execute the query
    cursor.execute(query)

    # Fetch and print results
    results = cursor.fetchall()
    #print("results", results)
    for row in results:
        print(f"First Name: {row[0]}, Last Name: {row[1]}")
except mysql.connector.Error as err:
    print(f"Error: {err}")
```
